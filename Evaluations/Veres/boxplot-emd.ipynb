```python
import os
os.chdir('JQ/PI-SDE')
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
leaveout_time = 'leaveout7'
leaveout_t = ['t7']
```

```python
df = pd.DataFrame(columns=['eval', 'time', 'methods','EMD'])

path = '../OtherMethods/TrajectoryNet/results/Veres/' + leaveout_time +'/emd.log'
method = 'TrajectoryNet'  
log = pd.read_csv(path, sep = '\t')
for i in range(5):  # seed
    for time in ['1','2','3','4','5','6','7']:
        if len(leaveout_t)==1:
            if ('t'+time) == leaveout_t[0]:
                df = df.append({'eval':'test', 'time': 'Day ' + time, 'methods':method,
                                'EMD':log.iloc[i]['t'+time] },ignore_index=True)
            else:
                df = df.append({'eval':'train', 'time': 'Day ' + time, 'methods':method,
                                'EMD':log.iloc[i]['t'+time] },ignore_index=True)
        else:
            if ('t'+time) in leaveout_t:
                df = df.append({'eval':'test', 'time': 'Day ' + time, 'methods':method,
                                'EMD':log.iloc[i]['t'+time] },ignore_index=True)
            else:
                df = df.append({'eval':'train', 'time': 'Day ' + time, 'methods':method,
                                'EMD':log.iloc[i]['t'+time] },ignore_index=True)
                

path = '../OtherMethods/MIOFlow/results/Veres/0.0002/' + leaveout_time +'/emd.log'
method = 'MIOFlow'  
log = pd.read_csv(path, sep = '\t')
for i in range(5):  # seed
    for time in ['1','2','3','4','5','6','7']:
        if len(leaveout_t)==1:
            if ('t'+time) == leaveout_t[0]:
                df = df.append({'eval':'test', 'time': 'Day ' + time, 'methods':method,
                                'EMD':log.iloc[i]['t'+time] },ignore_index=True)
            else:
                df = df.append({'eval':'train', 'time': 'Day ' + time, 'methods':method,
                                'EMD':log.iloc[i]['t'+time] },ignore_index=True)
        else:
            if ('t'+time) in leaveout_t:
                df = df.append({'eval':'test', 'time': 'Day ' + time, 'methods':method,
                                'EMD':log.iloc[i]['t'+time] },ignore_index=True)
            else:
                df = df.append({'eval':'train', 'time': 'Day ' + time, 'methods':method,
                                'EMD':log.iloc[i]['t'+time] },ignore_index=True)
                

path = '../OtherMethods/PRESCIENT/RESULTS/Veres/' + leaveout_time +'/kegg/emd-best.log'
method = 'PRESCIENT(+growth)'  
log = pd.read_csv(path, sep = '\t')
for i in range(5):  # seed
    for time in ['1','2','3','4','5','6','7']:
        if len(leaveout_t) == 1:
            if ('t'+time) == leaveout_t[0]:
                df = df.append({'eval':'test', 'time': 'Day ' + time, 'methods':method,
                                'EMD':log.iloc[i]['t'+time] },ignore_index=True)
            else:
                df = df.append({'eval':'train', 'time': 'Day ' + time, 'methods':method,
                                'EMD':log.iloc[i]['t'+time] },ignore_index=True)
        else:
            if ('t'+time) in leaveout_t:
                df = df.append({'eval':'test', 'time': 'Day ' + time, 'methods':method,
                                'EMD':log.iloc[i]['t'+time] },ignore_index=True)
            else:
                df = df.append({'eval':'train', 'time': 'Day ' + time, 'methods':method,
                                'EMD':log.iloc[i]['t'+time] },ignore_index=True)
                            
path = '../OtherMethods/PRESCIENT/RESULTS/Veres/' + leaveout_time +'/none/emd-best.log'
method = 'PRESCIENT'  
log = pd.read_csv(path, sep = '\t')
for i in range(5):  # seed
    for time in ['1','2','3','4','5','6','7']:
        if len(leaveout_t) == 1:
            if ('t'+time) == leaveout_t[0]:
                df = df.append({'eval':'test', 'time': 'Day ' + time, 'methods':method,
                                'EMD':log.iloc[i]['t'+time] },ignore_index=True)
            else:
                df = df.append({'eval':'train', 'time': 'Day ' + time, 'methods':method,
                                'EMD':log.iloc[i]['t'+time] },ignore_index=True)
        else:
            if ('t'+time) in leaveout_t:
                df = df.append({'eval':'test', 'time': 'Day ' + time, 'methods':method,
                                'EMD':log.iloc[i]['t'+time] },ignore_index=True)
            else:
                df = df.append({'eval':'train', 'time': 'Day ' + time, 'methods':method,
                                'EMD':log.iloc[i]['t'+time] },ignore_index=True)
                
                
for seedd in ['seed_0/','seed_1/','seed_2/','seed_3/','seed_4/']:
    path = 'RESULTS/Veres/none-softplus-400_400-0.5-const-0.1-0.1-0.005/' + leaveout_time + seedd + '/interpolate-emd.log'
    method = 'PI-SDE' 
    log = pd.read_csv(path, sep = '\t')
    last = log.loc[log['epoch'] == 'best']
    
    for time in [1,2,3,4,5,6,7]:
        if len(leaveout_t) == 1:
            if str(time) == leaveout_t[0].split('t')[1]:
                df = df.append({'eval':'test', 'time': 'Day ' + str(time), 'methods':method,
                                'EMD':last[last['t'] == time]['loss'].item() },ignore_index=True)
            else:
                df = df.append({'eval':'train', 'time': 'Day ' + str(time), 'methods':method,
                                'EMD':last[last['t'] == time]['loss'].item() },ignore_index=True)
        else:
            leaveout_tt= [s.split('t')[1] for s in leaveout_t]
            if str(time) in leaveout_tt:
                df = df.append({'eval':'test', 'time': 'Day ' + str(time), 'methods':method,
                                'EMD':last[last['t'] == time]['loss'].item() },ignore_index=True)
            else:
                df = df.append({'eval':'train', 'time': 'Day ' + str(time), 'methods':method,
                                'EMD':last[last['t'] == time]['loss'].item() },ignore_index=True)
                

sns.set(style="whitegrid")

if len(leaveout_t)==1:
    fig, axs = plt.subplots(1, 2, figsize=(10, 6), gridspec_kw={'width_ratios': [6, 1.1]})  
elif len(leaveout_t)==2:
    fig, axs = plt.subplots(1, 2, figsize=(10, 6), gridspec_kw={'width_ratios': [5, 2.1]})

    
df_train = df[df['eval']=='train']
sns.barplot(x="time", y="EMD", hue="methods", data=df_train, palette='Set2',  errorbar='sd', ax=axs[0])
sns.stripplot(x="time", y="EMD", hue="methods", data=df_train, palette='Set2',  alpha=0.5, ax=axs[0], dodge=True, legend=False)
axs[0].set_xlabel('Train t', fontsize=20)
axs[0].set_ylabel('') 
axs[0].set_xticklabels(axs[0].get_xticklabels(), fontsize=20)

df_test = df[df['eval']=='test']
sns.barplot(x="time", y="EMD", hue="methods", data=df_test, palette='Set2',  errorbar='sd', ax=axs[1])
sns.stripplot(x="time", y="EMD", hue="methods", data=df_test, palette='Set2',  alpha=0.5, ax=axs[1], dodge=True, legend=False)
axs[1].set_xlabel('Test t', fontsize=20)
axs[1].set_ylabel('') 
axs[1].set_xticklabels(axs[1].get_xticklabels(), fontsize=20)


fig.text(-0.01, 0.5, 'Earth Mover Distance', va='center', rotation='vertical',fontsize=20)

for ax in axs:
    ax.get_legend().remove()

handles, labels = axs[0].get_legend_handles_labels()
fig.legend(handles, labels, loc='center right', bbox_to_anchor=(1.27, 0.5), ncol=1)

plt.tight_layout()
plt.show()

```
